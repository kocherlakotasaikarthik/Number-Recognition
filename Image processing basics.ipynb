**Thresholding**

** Find values greater than 127 in an array "gray" , replace that with max value and if it is less that 127 replace with min value**

```python
cond1 = gray>50
cond2 = gray<50
gray[cond2] = gray.max() # finding values less than 127, then repalce with max()
gray[cond1] = gray.max() # finding values greater than 127, then repalce with min()
```

```python
gray_norm  = gray/gray.max()  # 0 - 1 -> normalisation
```

# Working with video 

```python
cap = cv2.VideoCapture(0)

while True:
    ret, img = cap.read() # read output of your webcam
    
    cv2.imshow('video',img) # window will open
    if cv2.waitKey(1) == 27: # esc to close
        break
        
cap.release()
cv2.destroyAllWindows()
```
